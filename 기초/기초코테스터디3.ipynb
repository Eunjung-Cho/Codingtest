# 1. 파이썬 코딩 도장 28.4 심사문제: 파일에서 회문인 단어 출력하기

- [https://dojang.io/mod/quiz/view.php?id=2334](https://dojang.io/mod/quiz/view.php?id=2334)

1. 먼저 open으로 words.txt 파일을 읽기 모드 'r'로 엽니다. 
 
2. 그다음에 readlines로 파일의 내용을 한 줄씩 읽어서 리스트 형태로 가져옵니다(리스트는 이하 words).

3. 파일에서 읽은 단어는 \n이 붙어있으므로 \n을 제외해야 합니다. 따라서 word = word.strip('\n')과 같이 \n을 삭제하여 다시 word에 저장합니다. 

4. 회문 판단 방법: 반복문, 시퀀스 뒤집기, 리스트와 reversed, 문자열의 join 메서드와 reversed를 사용하는 방법 들이 있는데 여기서 reversed 함수 사용






```python
with open('words.txt','r')as file: # words.txt파일 불러오기
    for line in file:
        words=line.split() #공백을 기준으로 분리하여 리스트로
        for word in words:
            if list(word)==list(reversed(word)): #만약 회문이면 출력
                print(word.strip())
```
